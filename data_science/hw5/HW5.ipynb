Now, turn these data into a NetworkX graph, according to the spec below. For details on using NetworkX, consult the lab materials for November 1, as well as the [NetworkX documentation](http://networkx.github.io/).

```python
"""
Function
--------
vote_graph

Parameters
----------
data : list of dicts
    The vote database returned from get_vote_data

Returns
-------
graph : NetworkX Graph object, with the following properties
    1. Each node in the graph is labeled using the `display_name` of a Senator (e.g., 'Lee (R-UT)')
    2. Each node has a `color` attribute set to 'r' for Republicans, 
       'b' for Democrats, and 'k' for Independent/other parties.
    3. The edges between two nodes are weighted by the number of 
       times two senators have cast the same Yea or Nay vote
    4. Each edge also has a `difference` attribute, which is set to `1 / weight`.

Examples
--------
>>> graph = vote_graph(vote_data)
>>> graph.node['Lee (R-UT)']
{'color': 'r'}  # attributes for this senator
>>> len(graph['Lee (R-UT)']) # connections to other senators
101
>>> graph['Lee (R-UT)']['Baldwin (D-WI)']  # edge relationship between Lee and Baldwin
{'difference': 0.02, 'weight': 50}
"""
# **********
# Vote graph main method and helpers
#
# create_voting_history: Assigns all nays and yeas to each senator
# links_senators: Links senators based on common yes and no
# get_node_color: Reurns the node color based on party affliation
#
# vote_graph: Returns the resultant graphs for these senators 
#
# **********

# Iterate through all the bills and create voting history for each senator.
# key is the senator display name - values have two groups yeas and nays containing bill numbers
def create_voting_history(data):
    senators = {}
    for vote in data:
        
        bill_num = vote['number']
        yeas = vote['votes']['Yea']
        nays = vote['votes']['Nay']
        for yea in yeas:
            display_name= yea['display_name']
            party = yea['party'] 
            if (display_name, party) in senators.keys():
                (my_yeas, my_nays) = senators[(display_name, party)]
            else:
                (my_yeas, my_nays) = ([], [])
            my_yeas.append(bill_num)
            senators[(display_name, party)] = (my_yeas, my_nays)   
            
        for nay in nays:
            display_name= nay['display_name']
            party = nay['party'] 
            if (display_name, party) in senators.keys():
                (my_yeas, my_nays) = senators[(display_name, party)]
            else:
                (my_yeas, my_nays) = ([], [])
            my_nays.append(bill_num)
            senators[(display_name, party)] = (my_yeas, my_nays)   
    return senators


# Link senators based on common yeas and nays on a vote
def link_senators(this_senator_info, senators, bipartisan=False):
    linked_senators = {}
    
    (name, party, yeas, nays) = this_senator_info     
    
    #print "name=", name, "party=", party
    for key_tuple, value_tuple in senators.iteritems():
        (o_name, o_party) = key_tuple
        (o_yeas, o_nays) = value_tuple
        if o_name == name:
            continue
        if bipartisan == True:
            if party == o_party:
                continue
        yeas_common = set(sorted(yeas)).intersection(sorted(o_yeas))
        nays_common = set(sorted(nays)).intersection(sorted(o_nays))  
        
        #print name, "->", o_name, "yeas=", len(yeas_common), "nays=", len(nays_common)
        
        total_common = len(yeas_common) + len(nays_common)     
        linked_senators[key_tuple] = total_common
 
    return linked_senators


def xget_node_color(party):
    if party == 'R':
        return 'r'
    elif party == 'D':
        return 'b'
    else:
        return 'k'

def get_node_color(sen_name):
    if 'R-' in sen_name:
        return 'r'
    elif 'D-' in sen_name:
        return 'b'
    else:
        return 'k'

# The graph object based on senators yeas and nays on a common vote
# The number of common yeas or nays constitutes the weight of the link.
# An inverse "difference" is also added (later used for min. spanning)
def vote_graph(data, bipartisan=False):
    
    G = nx.Graph()
    voting_hist = create_voting_history(data)
    
    # set up all the nodes without any links
    for key_tuple, value_tuple in voting_hist.iteritems():
        (name, party) = key_tuple
        (yeas, nays) = value_tuple
        party_color = get_node_color(name)
        
        G.add_node(name, color=party_color)
    
    # link the nodes together 
    for key_tuple, value_tuple in voting_hist.iteritems():
        (name, party) = key_tuple
        (yeas, nays) = value_tuple
        senator_info = (name, party, yeas, nays)
        other_senators = link_senators(senator_info, voting_hist, bipartisan)
        for o_key, o_value in other_senators.iteritems():
            o_name, o_party = o_key
            if o_value > 0:
                G.add_edge(name, o_name, weight=o_value) 
                G[name][o_name]['difference'] = float(1/float(o_value))
    return G


# ********** Test Methods **********
def test_create_voting_history():
    print ""
    print "** test_create_voting_history"
    print ""
    test_senators = create_voting_history(vote_data)
    print "number of senators", len(test_senators.keys())
    test_yeas = test_senators[('Flake (R-AZ)', 'R')][0]
    test_nays = test_senators[('Flake (R-AZ)', 'R')][1]
    print "100 in test_yeas", 100 in test_yeas #True
    print "101 in test_yeas", 101 in test_yeas #False
    print "101 in test_nays", 101 in test_nays #True


def test_link_senators():
    print ""
    print "** test_link_senators"
    print ""
    test_senators = create_voting_history(vote_data)
    test_senator = test_senators[('Flake (R-AZ)', 'R')]
    test_yeas = test_senators[('Flake (R-AZ)', 'R')][0]
    test_nays = test_senators[('Flake (R-AZ)', 'R')][1]
    this_senator_info = ('Flake (R-AZ)', 'R', test_yeas, test_nays)
    
    test_linked_senators = link_senators(this_senator_info, test_senators)
    print len(test_linked_senators.keys())
    #print test_linked_senators


def test_vote_graph():
    print ""
    print "** test_vote_graph"
    print ""
    graph = vote_graph(vote_data)
    print graph.node['Lee (R-UT)'] # {'color': 'r'}
    print len(graph['Lee (R-UT)']) # 101
    print graph['Lee (R-UT)']['Baldwin (D-WI)'] # {'difference': 0.02, 'weight': 50}
    #Lee (R-UT) -> Baldwin (D-WI) yeas= 42 nays= 9

if TESTING:
    test_create_voting_history()
    test_link_senators()
    test_vote_graph()
    # expected
    # {'color': 'r'}
    # 102
    # {'difference': 0.0196078431372549, 'weight': 51}
```

### How (and how not) to visualize networks

Network plots often *look* impressive, but creating sensible network plots is tricky. From Ben Fry, the author of the Processing program:
<blockquote>
Usually a graph layout isn’t the best option for data sets larger than a few dozen nodes. You’re most likely to wind up with enormous spider webs or balls of string, and the mess seen so far is more often the case than not. Graphs can be a powerful way to represent relationships between data, but they are also a very abstract concept, which means that they run the danger of meaning something only to the creator of the graph. Often, simply showing the structure of the data says very little about what it actually means, even though it’s a perfectly accurate means of representing the data. Everything looks like a graph, but almost nothing should ever be drawn as one.
</blockquote>

Let's look at bad and better ways of visualizing the senate vote network.

First, consider the "default" plot from networkx.

```python
#this makes sure draw_spring results are the same at each call
np.random.seed(1)  

color = [votes.node[senator]['color'] for senator in votes.nodes()]

#determine position of each node using a spring layout
pos = nx.spring_layout(votes, iterations=200)

#plot the edges
nx.draw_networkx_edges(votes, pos, alpha = .02) #default is a complete mess

#plot the nodes
nx.draw_networkx_nodes(votes, pos, node_color=color)

#draw the labels
lbls = nx.draw_networkx_labels(votes, pos, font_size=8)

#coordinate information is meaningless here, so let's remove it
plt.xticks([])
plt.yticks([])
remove_border(left=False, bottom=False)
```

The spring layout tries to group nodes with large edge-weights near to each other. In this context, that means it tries to organize the Senate into similarly-voting cliques. However, there's simply too much going on in this plot -- we should simplify the representation.

Compute the `Minimum Spanning Tree` of this graph, using the `difference` edge attribute as the weight to minimize. A [Minimum Spanning Tree](http://en.wikipedia.org/wiki/Minimum_spanning_tree) is the subset of edges which trace at least one path through all nodes ("spanning"), with minimum total edge weight. You can think of it as a simplification of a network.

Plot this new network, making modifications as necessary to prevent the graph from becoming too busy.

```python
#Your code here
# A couple of implementations are provided - for smaller resolution and large
# The best resolution is turned on(draw_large)
# The smaller version can be rendered on by uncommenting the calling method
def draw_large():
    fig = plt.figure(figsize=(30, 20))
    
    votes_mst = nx.minimum_spanning_tree(votes, weight='difference')
    color = [votes.node[senator]['color'] for senator in votes_mst.nodes()]
    
    pos = nx.spring_layout(votes_mst, iterations=200, scale=200)
    nx.draw_networkx_nodes(votes_mst, pos, node_color=color, node_size=150, alpha=0.7, linewidths=0.5)
    nx.draw_networkx_edges(votes_mst, pos, alpha = .5)
    nx.draw_networkx_labels(votes_mst, pos, alpha=1, font_size=10, font_color='k', font_weight='normal')
    
    #coordinate information is meaningless here, so let's remove it
    plt.xticks([])
    plt.yticks([])
    remove_border(left=False, bottom=False)

    
def draw_small():
    
    votes_mst = nx.minimum_spanning_tree(votes, weight='difference')
    color = [votes.node[senator]['color'] for senator in votes_mst.nodes()]
    
    pos = nx.spring_layout(votes_mst, iterations=200, scale=200)
    nx.draw_networkx_nodes(votes_mst, pos, node_color=color, node_size=100, alpha=0.3, linewidths=0.5)
    nx.draw_networkx_edges(votes_mst, pos, alpha = .5)
    nx.draw_networkx_labels(votes_mst, pos, alpha=1, font_size=5, font_color='k', font_weight='bold')
    
    #coordinate information is meaningless here, so let's remove it
    plt.xticks([])
    plt.yticks([])
    remove_border(left=False, bottom=False)


# PLEASE SWAP IT IF RESOLUTION BECOMES AN ISSUE
#draw_small()
draw_large()
```

###Minimum Spanning Tree

**Definition:** Networkx implmentation uses Kruskal's algorithm that finds a minimum spanning tree for a connected weighted graph. This means it finds a subset of the edges that forms a tree that includes every vertex, where the total weight of all the edges in the tree is minimized [http://en.wikipedia.org/wiki/Kruskal's_algorithm]. 


**In Our Context:** We are using "difference" as a contributing factor to connect the minimum spanning tree. Difference is the inverse of the weights that we calculated - smaller the difference the greater the cooperation between the senators. The minimum spanning tree will take the smallest costing routes (smallest differences) and will end up connecting the senators with the greatest cooperation.

- To create the min. spanning tree, bi-partisans will be pushed closer to the center of the graph so that they can be used in more connections.


- The algorithm will try to re-use the connections with the smallest differences when connecting the senators. The result will be that the most bi-partisans will be pushed on the main trunk.


- The leaf nodes present a curious case. My interpretation is that when there is a tie or the nodes are nearly equally bi-partisans but not all of them can be put on the trunk, the algorithm tags them as leaves.


**Most bi-partisan**

Most bi-partisans will be closest to each other when the links for the two groups are connected. Within each group, most probably, you would have the same weight as everyone votes together. ##Discussion

###Closeness centrality in general

Closeness centrality scores define how close a node is to the rest of the nodes. Quoting wikipedia, *"the more central a node is the lower its total distance to all other nodes. Closeness can be regarded as a measure of how long it will take to spread information from this to all other nodes sequentially"*. In essence, the closeness centrality is the sum of the minimum distance of all nodes from a given node. 


**Networkx Implementation:** 
Closeness centrality at a node is 1/average distance to all other nodes.

$$C_H(x)= \sum_{y \neq x}\frac{1}{d(y,x)}$$

The closeness centrality is normalized to to $\frac{n-1}{size(G)-1}$ where n is the number of nodes in the connected part of graph containing the node. If the graph is not completely connected, this algorithm computes the closeness centrality for each connected part separately.


###Closeness Centrality in our case

- **Centrality measure:** We are using "difference" as the centrality measure which is the inverse of weights assigned when two senators agree with each other (yes or no). 

- **Normalization**: Normalization probably has no effect in our case since the graph seems to be completely connected and there do not seem to be small islands.

## Leadership in the Senate

There are many metrics to quantify the leadership in the Senate.

  1. Senate leaders sponsor and co-sponsor lots of bills
  2. Leaders sit on many committees, as well as more important committees
  3. Leaders usually have been in office for a long time
  
Another approach uses the philosophy behind how Google ranks search results. The core idea behind Google's PageRank algorithm is:

1. A "good" website (i.e. one to rank highly in search results) is linked to by many other websites
2. A link found on a "good" website is more important than a link found on a "bad" website

The PageRank algorithm thus assigns scores to nodes in a graph based on how many neighbors a node has, as well as the score of those neighbors.

This technique can be adapted to rank Senate leadership. Here, nodes correspond to Senators, and edges correspond to a senator co-sponsoring a bill sponsored by another Senator. The weight of each edge from node A to B is the number of times Senator A has co-sponsored a bill whose primary sponsor is Senator B. If you interpret the PageRank scores of such a network to indicate Senate leadership, you are then assuming:

1. Leaders sponsor more bills
1. Leaders attract co-sponsorship from other leaders

### Problem 8

Write a function to builded a Directed Graph (DiGraph) from these data, according to the following spec:

```python
"""
Function
--------
bill_graph

Turn the bill graph data into a NetworkX Digraph

Parameters
----------
data : list of dicts
    The data returned from get_all_bills

Returns
-------
graph : A NetworkX DiGraph, with the following properties
    * Each node is a senator. For a label, use the 'name' field 
      from the 'sponsor' and 'cosponsors' dict items
    * Each edge from A to B is assigned a weight equal to how many 
      bills are sponsored by B and co-sponsored by A
"""
#Your code here

# The main method bill_graph is broknen into the following
#
# get_senators uses the list of bills 

# create a dictionary with key as the sponsor
# value is a dictionary with cosponsors and the count they have consponsored
# The code is a little tricky
# We loop through the bills and pick up the sponsors and related cosponsors for that bill
# We check if the sponsor is already in teh sponsor dictionary or add it
# We now loop through the list of cosponsors and see if the cosponsor is in the list
# If not we add him/her, otherwise we increase the count
def get_senators(data):
    sponsors = {}
    for bill in data:
        sponsor= bill['sponsor']['name']
        cosponsers = bill['cosponsors']
        
        cosponser = None
        for cosponser_info in cosponsers:
            cosponser = cosponser_info['name']
            cosponser_dict = {}
            if sponsor in sponsors.keys():
                cosponser_dict = sponsors[sponsor]
            else:
                cosponser_dict = {}
                sponsors[sponsor] = cosponser_dict
        
            if cosponser in cosponser_dict.keys():
                cosponser_dict[cosponser] = cosponser_dict[cosponser] + 1
            else:
                try:
                    cosponser_dict[cosponser] = 1
                except:
                    pass

            sponsors[sponsor] = cosponser_dict
    return sponsors


# Each edge from A to B is assigned a weight equal to how many 
# bills are sponsored by B and co-sponsored by A
def get_edge_A_to_B(sponsor_dict, senA, senB):
    
    count = 0
    cosponsors = sponsor_dict[senB]
    cosponsors_keys = cosponsors.keys()
    #print senB, cosponsors.keys()
    if senA in cosponsors_keys:
        count = cosponsors[senA]
    
    return count

        
    
def bill_graph(data):
    G = nx.DiGraph()
    senators = get_senators(data)
    s_keys = senators.keys()
    
    # first create the graphs with all senators
    for s_key in s_keys:
        G.add_node(s_key)
        #print s_key
    
    # now link the graphs with the weight
    for senA in s_keys:
        for senB in s_keys:
            if senA == senB:
                continue
            weight_sponsor = get_edge_A_to_B(senators, senA, senB)
            
            if weight_sponsor > 0:
                G.add_edge(senA, senB, weight=weight_sponsor)
                #G.add_edge(1, 3, capacity=15, length=342.7)                      
    return G


# ********** TESTING **********

def test_get_senators():
    print ""
    print "** test_get_senators"
    print ""
    test_senators = get_senators(bill_list)
    test_senators_keys = test_senators.keys()
    print test_senators_keys[0], test_senators[test_senators_keys[0]]
    
    
    #Burr, Richard and Hagan, Kay have 3 connections
    test_sponsor = 'Burr, Richard'
    test_cosponsor = 'Hagan, Kay' #'Crapo'#
    test_cosponsors = 0
    for bill in bill_list:
        if bill['sponsor']['name'] == test_sponsor:
            for co in bill['cosponsors']:
                #print co['name']
                if  test_cosponsor in co['name']:
                    test_cosponsors += 1
    print ""
    print "** Checking connections between", test_sponsor,"and", test_cosponsor
    print test_sponsor, "->", test_cosponsor, "[ connections:", test_cosponsors, "]"
    print ""

if TESTING:
    pass
    test_get_senators()
```
